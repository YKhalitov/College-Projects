##Features and Target Isolation

```python
y = Data['Survived']
X = Data.drop(columns = ['Survived', 'PassengerId', 'Name', 'Ticket','Cabin', "Embarked"]) #cut out things we did not need, goal is to get to numbers
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.head()
```

## Model Selection


```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train, y_train) #what to do in homework, important to fit on train data and not on entirety of set
```

```python
y_pred = neigh.predict(X_test)
print(y_pred)
```

```python
sum(y_test == y_pred)/len(y_test) #ACCURACY
```
